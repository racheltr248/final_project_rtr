```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import clean licensors csv
url = 'https://raw.githubusercontent.com/Megreid23/final_project/main/csv%20files/licensors.csv'
licensors = pd.read_csv(url)
licensors.head()
```

```python
licensors.drop(columns = ['Unnamed: 0'], inplace=True)
licensors.head()
```

```python
# convert list-like object columns into lists
# https://towardsdatascience.com/dealing-with-list-values-in-pandas-dataframes-a177e534f173
licensors['licensors'] = licensors['licensors'].apply(eval)
licensors.head()
```

```python
# separate list-like column into separate columns
# https://stackoverflow.com/questions/50264678/explode-column-of-lists-into-multiple-columns
series = licensors['licensors'].apply(pd.Series)
series.head()
```

```python
# merge licensors table and series
# https://stackoverflow.com/questions/26265819/how-to-merge-a-series-and-dataframe/40762674#40762674
licensors_merged = licensors.merge(series, left_index=True, right_index=True)
licensors_merged
```

```python
# drop original list column
licensors_merged.drop(columns = ['licensors'], inplace=True)
licensors_merged.head()
```

```python
import pandas as pd
```

```python
# import clean demographics csv (RTR)
url = 'https://raw.githubusercontent.com/Megreid23/final_project/main/csv%20files/demographics.csv'
demographics = pd.read_csv(url)
demographics
```

```python
demographics.drop(columns = ['Unnamed: 0'], inplace=True)
demographics
```

```python
# ''.join(demographics['demographics'].replace("[","").replace("]", ""))
# demographics
```

```python
# convert list-like object columns into lists
# https://towardsdatascience.com/dealing-with-list-values-in-pandas-dataframes-a177e534f173
demographics['demographics'] = demographics['demographics'].apply(eval)
demographics
```

```python
demographics.shape
```

```python
demographics.info()
```

```python
# separate list-like column into separate columns
# https://stackoverflow.com/questions/50264678/explode-column-of-lists-into-multiple-columns
series = demographics['demographics'].apply(pd.Series)
series
```

```python
# merge demographics table and series
# https://stackoverflow.com/questions/26265819/how-to-merge-a-series-and-dataframe/40762674#40762674
demographics_merged = demographics.merge(series, left_index=True, right_index=True)
demographics_merged
```

```python
demographics_merged.head()
```

```python
demographics_merged.drop(columns = ['demographics'], inplace=True)
demographics_merged.head()
```
